```python
from insightface.app import FaceAnalysis
app = FaceAnalysis(
    name="antelopev2",
    root="/content",
    providers=["CUDAExecutionProvider", "CPUExecutionProvider"],
    allowed_modules=["detection", "recognition"],
)
app.prepare(ctx_id=0, det_size=(640, 640))
print("Detection model:", app.models["detection"].model_file)
print("Recognition model:", app.models["recognition"].model_file)
```

### Датасет и split

```python

import kagglehub, random, shutil
from pathlib import Path

path = kagglehub.dataset_download("tommykamaz/faces-dataset-small")
DATA_ROOT = Path(path)

OUT_ROOT = Path("/content/datasets/ffhq_sub")
train_dir = OUT_ROOT / "train"
val_dir = OUT_ROOT / "val"
train_dir.mkdir(parents=True, exist_ok=True)
val_dir.mkdir(parents=True, exist_ok=True)

exts = (".png", ".jpg", ".jpeg", ".webp")
imgs = [p for p in DATA_ROOT.rglob("*") if p.suffix.lower() in exts]
assert imgs, f"No images in {DATA_ROOT}"

random.seed(42)
random.shuffle(imgs)
n_val = int(len(imgs) * 0.1)

for p in imgs[n_val:]:
    dst = train_dir / p.name
    if not dst.exists():
        shutil.copy2(p, dst)

for p in imgs[:n_val]:
    dst = val_dir / p.name
    if not dst.exists():
        shutil.copy2(p, dst)

print("train:", len(list(train_dir.iterdir())))
print("val:", len(list(val_dir.iterdir())))
```
